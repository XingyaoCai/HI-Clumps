```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1000
np.random.seed(42)
particles = np.random.rand(1000, 3) * 10  # 

# PCA
pca = PCA(n_components=2)
reduced_particles = pca.fit_transform(particles)

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(particles[:, 0], particles[:, 1], alpha=0.5)
ax1.set_title('Original Data')

ax2.scatter(reduced_particles[:, 0], reduced_particles[:, 1], alpha=0.5)
ax2.set_title('Reduced Dimensionality using PCA')

plt.show()
```
