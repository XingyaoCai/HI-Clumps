```python
import illustris_python as il
import numpy as np
import matplotlib.pyplot as plt
```

```python
basePath = "./sims.TNG/TNG50-1/output"
Current_Snapshot = 99
Subhalos_Dict_TNG50 = il.groupcat.loadSubhalos(
    basePath, Current_Snapshot, fields=['SubhaloVmaxRad', 'SubhaloMass'])
condition = Subhalos_Dict_TNG50['SubhaloMass'] > 100
indices = np.where(condition)[0]
print(f'Number of Galaxies {indices.shape[0]}')
SubhaloVmaxRad_Array_TNG50 = Subhalos_Dict_TNG50['SubhaloVmaxRad'][indices]
print(f'Mean={SubhaloVmaxRad_Array_TNG50.mean()}\nMax={SubhaloVmaxRad_Array_TNG50.max()}')
plt.hist(SubhaloVmaxRad_Array_TNG50, bins=10)
plt.xlabel('SubhaloVmaxRad [ckpc/h]')
plt.ylabel('Subhalos Num')
plt.title('Distribution of TNG50')
```

```python

SubhaloVmaxRad_Array_TNG50 = Subhalos_Dict_TNG50['SubhaloVmaxRad'][indices]
plt.hist(SubhaloVmaxRad_Array_TNG50, bins=10)
plt.xlabel('SubhaloVmaxRad [ckpc/h]')
plt.ylabel('Subhalos Num')
plt.yscale('log')
plt.title('Distribution of TNG50')
```

```python
basePath = "./sims.TNG/TNG100-1/output"
Current_Snapshot = 99
Subhalos_Dict = il.groupcat.loadSubhalos(basePath, Current_Snapshot, fields=[
                                         'SubhaloVmaxRad', 'SubhaloMass'])
condition = Subhalos_Dict['SubhaloMass'] > 100
indices = np.where(condition)[0]
print(f'Number of Galaxies {indices.shape[0]}')
SubhaloVmaxRad_Array = Subhalos_Dict['SubhaloVmaxRad'][indices]
print(f'Mean={SubhaloVmaxRad_Array.mean()}\nMax={SubhaloVmaxRad_Array.max()}')
plt.hist(SubhaloVmaxRad_Array, bins=10, range=([0, 200]))
plt.xlabel('SubhaloVmaxRad [ckpc/h]')
plt.ylabel('Subhalos Num')
plt.yscale('log')
plt.title('Distribution of TNG100')
```
